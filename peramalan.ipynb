```python
import pandas as pd
df = "airline-passengers.csv"
data =pd.read_csv(df)
data.tail()
```

```python
# from sklearn.model_selection import train_test_split
# X_train, X_test=train_test_split(data, test_size=0.2, random_state=1)
# X_train.shape + X_test.shape
```

```python
training_data = data.iloc[:115, 1:2].values
test_data = data.iloc[115:, 1:2]
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

scaler = MinMaxScaler(feature_range = (0,1))
training_set_scaled = scaler.fit_transform(training_data)

x_train = []
y_train = []
x = 3

for i in range(x, 115):
    x_train.append(training_set_scaled[i-x:i, 0])
    y_train.append(training_set_scaled[1,0])

x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))


x_train.shape
```

```python
x_train
```

```python
newX = pd.DataFrame(x_train.reshape(112,3),columns=['Xt-2','Xt-1','Xt'])
newY = pd.DataFrame(y_train, columns=['Output'])
```

```python
new_data_passenger = pd.concat([newX,newY],axis=1)
new_data_passenger  
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test = train_test_split(newX, train_size = 0.8, test_size = 0.2, shuffle=False)
y_train, y_test = train_test_split(newY, train_size = 0.8, test_size = 0.2, shuffle=False)
```

```python
# buat model knn regressor
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=1)

# train model
model.fit(X_train, y_train)

# evaluasi model
evaluasi = model.score(X_test, y_test)
print('Hasil Evaluasi :',evaluasi)

# prediksi data test
y_pred = model.predict(X_test)
print('Hasil prediksi :',y_pred)
```
